```python
import requests
from bs4 import BeautifulSoup

URL = "https://en.wikipedia.org/wiki/COVID-19_pandemic_in_Malaysia"

res = requests.get(URL).text
soup = BeautifulSoup(res,'lxml')
```

```python
#soup_list = soup.find('table', class_='wikitable sortable mw-collapsible mw-collapsed').tbody
```

```python
soup2_list = soup.findAll('table', {'class': 'wikitable sortable mw-collapsible mw-collapsed'})
for soup2 in soup2_list:
    row = soup2.findAll('tr')
```

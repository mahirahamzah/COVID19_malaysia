```python
import requests
from bs4 import BeautifulSoup

URL = "https://en.wikipedia.org/wiki/COVID-19_pandemic_in_Malaysia"

res = requests.get(URL).text
soup = BeautifulSoup(res,'lxml')
```

```python
soup_list = soup.find('table', class_='wikitable mw-datatable').tbody
```

```python
len(soup_list)
```

```python
row = soup_list.find_all('tr')
```

```python
column = [v.text.replace('\n', '') for v in row[0].find_all('th')]
```

```python
column
```

```python
rows = row[1:]
```

```python
rows
```

```python
row = [item.text.split('\n') for item in rows]
```

```python
row
```

```python
rows = [item[1:19] for item in row]
```

```python
len(rows[0])
```

```python
import pandas as pd
```

```python
df2 = pd.DataFrame(rows,columns=column)
```

```python
df2
```

```python
df2.to_csv('Cases_ByState.csv')
```
